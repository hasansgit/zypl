```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
logs
```

**Описание каждой колонки**
* Run_Id - Номер посылки 
* Dur - Время посылки с начала контеста в секундах
* Size - Размер исходного кода посылки в байтах
* IP - IP адресс с которого была совершена посылка
* User_Id - ID участника совершившего посылку
* User_Login - login участника совершившего посылку
* User_Name - Фамилия, Имя и населённый пункт участника совершившего посылку 
* Prob - название задачи
* Lang - язык программирования 
* Status - вердикт проверяющей системы
* Score - Баллы за посылку
* Test - Количество пройденных тестов

```python
logs_valcount = logs['Lang'].value_counts()
logs_valcount
```

```python
langs_names = logs_valcount.index
langs_count = logs_valcount.values
langs_names, langs_count
```

```python
plt.figure(figsize=(13,5))
plt.bar(langs_names, langs_count)
```

```python
status_ok_valcounts = logs[logs['Status']=='OK']['Lang'].value_counts()
ok_langs_names = status_ok_valcounts.index
ok_count = status_ok_valcounts.values
status_ok_valcounts
```

```python
plt.bar(ok_langs_names, ok_count, color='green')
```

```python
ok_langs_valcount = logs[logs['Lang'].isin(ok_langs_names)]['Lang'].value_counts()
lang_div_ok_count = ok_langs_valcount / ok_count
lang_div_ok_count
# langs_count
# ok_count
# ok_langs_valcount
```

```python
plt.bar(ok_langs_names, lang_div_ok_count, color='red')
```

```python
lang_div_ok_count.std()
```

```python
logs_stat_valcount = logs['Status'].value_counts()
logs_stat_valcount
```

```python
plt.bar(logs_stat_valcount.index, logs_stat_valcount.values)
```

```python
for stat in sorted(list(set(logs['Status']))):
    plt.title(stat)
    bar = logs['Prob'].value_counts() / logs[logs['Status']==stat]['Prob'].value_counts()
    plt.bar(bar.index, bar.values, color='red')
    plt.show()
```

```python
for stat in sorted(list(set(logs['Status']))):
    bar = logs['Prob'].value_counts() / logs[logs['Status']==stat]['Prob'].value_counts()
    print('*'*30)
    print(stat)
    print(bar)
    print('*'*30)
```
